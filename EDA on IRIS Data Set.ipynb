# 2D-Scatter Plot

Let's start analysing the above data set by plotting a 2-D Scatter Plot between the first two features i.e, 'sepal_length' & 'sepal_width'.

```python
iris.plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width')
plt.show()
```

## Observations

We cannot make any good insights from the single coloured points plotted between 'sepal_length' & 'sepal_width'.
We should colour out each species with a different colour.

```python
sns.set_style('whitegrid')
# height is the size of the plot
sns.FacetGrid(iris, hue ='species', height = 8) \
    .map(plt.scatter, "sepal_length", "sepal_width") \
    .add_legend()
plt.show()
```

From the above plot, we can clearly classify 'Setosa' species from the other two species. But, to classify the other two species we must use another different plot. The Pair-Plots given below will give us move insights.

# Pair-Plot

Many of the times in Data Science/Machine Learning we have to deal with n-dimensional(n>3) problems, dimensions are 
the features in the given problem. Since we humans cannot imagine dimensions more than 3-d, we try to gain insights from n-d 
problems by plotting Pair-Plots of the features. All combinations of feature pairs are plotted and we gain insights into the
data.

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris, hue = "species", height = 4)
plt.show()
```

## Observations

From the above Pair-Plots, we can see from the plot between 'petal_width & petal_length', at (2,3) location, all the three species can be easily classified using the following if-else model:

#### Model to classify Setosa:

 if petal_width>0.0 and petal_width<=0.7 and petal_length>0 and petal_length<=2.0: species = 'Setosa'

#### Model to classify Versicolor:

if petal_width>=1.0 and petal_width<=1.75 and petal_length>=3 and petal_length<=5.1: species = 'Versicolor'

#### Model to classify Virginica:

if petal_width>=1.3 and petal_width<=2.5 and petal_length>=4.5 and petal_length<=7.0:
    species = 'Versicolor'
